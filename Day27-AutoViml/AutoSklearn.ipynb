```
import pandas as pd
import numpy as np
from autoviml.Auto_ViML import Auto_ViML
```

```
df=pd.read_csv('titanic.csv')
df.head()
```

```
df=df.sample(int(df.shape[0]))
size=int(0.7*df.shape[0])
print(size)
```

```
train=df[:size]
test=df[size:]
print(train.shape)
print(test.shape)
```

```
model,features,trainm,testm=Auto_ViML(train,'Survived',test,
                                      sample_submission='',hyper_param="GS",
                                      feature_reduction=True,KMeans_Featurizer=True,
                                      Boosting_Flag=True,verbose=1)
```

```
from sklearn.metrics import classification_report,confusion_matrix
confusion_matrix(test['Survived'],testm['Survived_Bagging_Classifier_predictions'])
```
